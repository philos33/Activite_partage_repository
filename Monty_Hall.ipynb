```python
# Pour afficher les graphiques dans la continuité du code, 
# et non pas dans une fenêtre à part:
%matplotlib inline

# Pour faire les calculs avec des vecteurs et tableaux
# et ne plus utiliser de boucle dans le processus de calcul
import numpy as np

# Un Enum est une structure de données qui consiste en un 
# ensemble d'éléments nommés. Une variable de ce type peut
# avoir comme valeur un de ces éléments.
from enum import Enum

# Pour pouvoir afficher des graphiques:
import matplotlib.pyplot as plt

#Pour importer le module qui contient la fonction de simulation du jeu de Monty Hall
import module as mod
```

```python
# Ici nous définissons une sous-classe de Enum, qui contiendra 
# les stratégies possibles.
class Strategie(Enum):
    CHANGER = 1
    GARDER = 2
```

```python
print("En changeant de porte, le joueur a gagné {} sur 10000 parties."
      .format(np.sum(mod.play_all(Strategie.CHANGER, 10000))))
      
print("En gardant son choix initial, le joueur a gagné {} sur 10000 parties."
      .format(np.sum(mod.play_all(Strategie.GARDER, 10000))))
```

```python
# plot renvoie un objet, que l'on pourra manipuler plus tard pour
# personnaliser le graphique
plot = plt.plot(mod.play_all(Strategie.CHANGER, 10))
```

```python
plot = plt.plot(mod.play_all(Strategie.CHANGER, 10000))
```

```python
plot = plt.scatter(range(10000), mod.play_all(Strategie.CHANGER, 10000))
```

```python
plot = plt.bar([1,2],[np.sum(mod.play_all(Strategie.CHANGER, 10000)), 
               np.sum(mod.play_all(Strategie.GARDER, 10000))], 
        tick_label=["Changer","Garder"])
```

```python
gains_changer = []
gains_garder = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for tours in samples:
    gains_changer.append(mod.play_all(Strategie.CHANGER, tours))
    gains_garder.append(mod.play_all(Strategie.GARDER, tours))
```

```python
figure = plt.figure()
plot = plt.scatter(samples, [sum(x) for x in gains_changer])
plot = plt.scatter(samples, [sum(x) for x in gains_garder])
```
